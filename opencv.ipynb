#BLURRING

```python
img=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
figure_size=7
result=cv2.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img1,cv2.COLOR_HSV2RGB)),plt.title("ORIGINAL")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2RGB)),plt.title("BLURRED")
plt.xticks([]),plt.yticks([])
plt.show()
```

#ZOOM

```python
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

#ROTATE

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```
